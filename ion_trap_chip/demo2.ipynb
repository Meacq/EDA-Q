## Example 1: Default Configuration

First, let's generate a chip with default parameters to see the baseline.

````python
def generate_default_chip():
    """Generate chip with default electrode parameters"""
    print("=" * 60)
    print("Example 1: Default Configuration")
    print("=" * 60)
    print("All electrodes: width=0.060mm, gap=0.040mm, depth=0.1mm\n")
    
    # Create chip generator with default parameters
    chip_gen = IonTrapChipGenerator()
    chip = chip_gen.create_complete_ion_trap_chip()
    
    if chip:
        print("✓ Chip generated successfully")
        chip_gen.export_to_step("chip_example1_default.step")
        print("✓ Exported to: chip_example1_default.step")
        chip_gen.visualize_chip_with_points()
    else:
        print("✗ Failed to generate chip")

# Run example
generate_default_chip()
````

## Example 2: Single Electrode Customization

Customize the first electrode in the `left_top_left` region to be wider - useful for a loading electrode.

````python
def generate_single_custom_electrode():
    """Generate chip with one customized electrode"""
    print("\n" + "=" * 60)
    print("Example 2: Single Electrode Customization")
    print("=" * 60)
    
    # Define custom parameters
    custom_params = {
        ('left_top_left', 0): {
            'electrode_width': 0.080,
            'gap_width': 0.045,
            'electrode_depth': 0.15,
        },
    }
    
    print("Configuration:")
    print("  - Electrode 0 in 'left_top_left': width=0.080mm, gap=0.045mm, depth=0.15mm")
    print("  - All other electrodes: default parameters\n")
    
    # Create chip generator with custom parameters
    chip_gen = IonTrapChipGenerator(custom_electrode_params=custom_params)
    chip = chip_gen.create_complete_ion_trap_chip()
    
    if chip:
        print("✓ Chip generated successfully")
        print("  Effect: Electrode 0 is wider, subsequent electrodes shifted by 0.025mm")
        chip_gen.export_to_step("chip_example2_single_custom.step")
        print("✓ Exported to: chip_example2_single_custom.step")
        chip_gen.visualize_chip_with_points()
    else:
        print("✗ Failed to generate chip")

# Run example
generate_single_custom_electrode()
````

## Example 3: Gradient Width Electrode Chain

Create a chain of electrodes with gradually decreasing widths.

````python
def generate_gradient_electrodes():
    """Generate chip with gradient electrode widths"""
    print("\n" + "=" * 60)
    print("Example 3: Gradient Width Electrode Chain")
    print("=" * 60)
    
    # Define gradient parameters
    custom_params = {
        ('left_top_left', 0): {'electrode_width': 0.080},
        ('left_top_left', 1): {'electrode_width': 0.075},
        ('left_top_left', 2): {'electrode_width': 0.070},
        ('left_top_left', 3): {'electrode_width': 0.065},
        ('left_top_left', 4): {'electrode_width': 0.060},
        # Electrodes 5-7 use default 0.060mm
    }
    
    print("Configuration: Gradient from 0.080mm to 0.060mm")
    print("  Electrode 0: 0.080mm")
    print("  Electrode 1: 0.075mm")
    print("  Electrode 2: 0.070mm")
    print("  Electrode 3: 0.065mm")
    print("  Electrodes 4-7: 0.060mm (default)\n")
    
    # Calculate positions
    widths = [0.080, 0.075, 0.070, 0.065, 0.060, 0.060, 0.060, 0.060]
    gap = 0.040
    position = 0.0
    
    print("Electrode positions:")
    for i, w in enumerate(widths):
        center = position + w / 2
        print(f"  E{i}: width={w:.3f}mm, center at {center:.3f}mm")
        position += (w + gap)
    
    print(f"\nTotal region length: {position:.3f}mm (default: 0.800mm)\n")
    
    # Create chip generator with custom parameters
    chip_gen = IonTrapChipGenerator(custom_electrode_params=custom_params)
    chip = chip_gen.create_complete_ion_trap_chip()
    
    if chip:
        print("✓ Chip generated successfully")
        chip_gen.export_to_step("chip_example3_gradient.step")
        print("✓ Exported to: chip_example3_gradient.step")
        chip_gen.visualize_chip_with_points()
    else:
        print("✗ Failed to generate chip")

# Run example
generate_gradient_electrodes()
````

## Example 4: Symmetric Configuration Across Multiple Regions

Apply the same customization to multiple regions for symmetric design.

````python
def generate_symmetric_configuration():
    """Generate chip with symmetric electrode customization"""
    print("\n" + "=" * 60)
    print("Example 4: Symmetric Configuration")
    print("=" * 60)
    
    # Define symmetric parameters for all 4 horizontal regions
    regions = ['left_top_left', 'right_top_right', 'left_bottom_left', 'right_bottom_right']
    custom_params = {}
    
    for region in regions:
        # First electrode wider for loading
        custom_params[(region, 0)] = {
            'electrode_width': 0.080,
            'gap_width': 0.045,
        }
        # Last electrode wider for detection
        custom_params[(region, 7)] = {
            'electrode_width': 0.075,
            'gap_width': 0.045,
        }
    
    print("Configuration:")
    print("  Applied to regions: left_top_left, right_top_right,")
    print("                      left_bottom_left, right_bottom_right")
    print("  - Electrode 0: width=0.080mm, gap=0.045mm (loading)")
    print("  - Electrodes 1-6: default parameters")
    print("  - Electrode 7: width=0.075mm, gap=0.045mm (detection)\n")
    
    # Create chip generator with custom parameters
    chip_gen = IonTrapChipGenerator(custom_electrode_params=custom_params)
    chip = chip_gen.create_complete_ion_trap_chip()
    
    if chip:
        print("✓ Chip generated successfully")
        print("  Effect: 4-fold rotational symmetry maintained")
        chip_gen.export_to_step("chip_example4_symmetric.step")
        print("✓ Exported to: chip_example4_symmetric.step")
        chip_gen.visualize_chip_with_points()
    else:
        print("✗ Failed to generate chip")

# Run example
generate_symmetric_configuration()
````

## Example 5: Complex Multi-Parameter Customization

Demonstrate customizing multiple parameters (width, gap, depth) for different electrodes with different functions.

````python
def generate_complex_customization():
    """Generate chip with complex multi-parameter customization"""
    print("\n" + "=" * 60)
    print("Example 5: Complex Multi-Parameter Customization")
    print("=" * 60)
    
    # Define complex parameters with different functions
    custom_params = {
        # Loading electrode - wide and deep
        ('left_top_left', 0): {
            'electrode_width': 0.100,
            'gap_width': 0.060,
            'electrode_depth': 0.15,
        },
        # RF electrodes - standard width, shallow depth
        ('left_top_left', 1): {'electrode_depth': 0.08},
        ('left_top_left', 2): {'electrode_depth': 0.08},
        ('left_top_left', 3): {'electrode_depth': 0.08},
        # Control electrodes - narrow with tight gaps
        ('left_top_left', 4): {
            'electrode_width': 0.050,
            'gap_width': 0.030,
        },
        ('left_top_left', 5): {
            'electrode_width': 0.050,
            'gap_width': 0.030,
        },
        # Detection electrode - wide
        ('left_top_left', 7): {
            'electrode_width': 0.090,
            'gap_width': 0.050,
        },
    }
    
    print("Configuration with functional specialization:")
    print("  E0: Loading    - w=0.100mm, g=0.060mm, d=0.15mm")
    print("  E1-3: RF       - default width/gap, d=0.08mm")
    print("  E4-5: Control  - w=0.050mm, g=0.030mm")
    print("  E6: Standard   - default parameters")
    print("  E7: Detection  - w=0.090mm, g=0.050mm\n")
    
    # Calculate positions
    configs = [
        (0, 0.100, 0.060, "Loading"),
        (1, 0.060, 0.040, "RF"),
        (2, 0.060, 0.040, "RF"),
        (3, 0.060, 0.040, "RF"),
        (4, 0.050, 0.030, "Control"),
        (5, 0.050, 0.030, "Control"),
        (6, 0.060, 0.040, "Standard"),
        (7, 0.090, 0.050, "Detection"),
    ]
    
    position = 0.0
    print("Electrode positions:")
    for idx, width, gap, func in configs:
        center = position + width / 2
        print(f"  E{idx} ({func:9s}): w={width:.3f}, g={gap:.3f}, center={center:.3f}mm")
        position += (width + gap)
    
    print(f"\nTotal region length: {position:.3f}mm\n")
    
    # Create chip generator with custom parameters
    chip_gen = IonTrapChipGenerator(custom_electrode_params=custom_params)
    chip = chip_gen.create_complete_ion_trap_chip()
    
    if chip:
        print("✓ Chip generated successfully")
        chip_gen.export_to_step("chip_example5_complex.step")
        print("✓ Exported to: chip_example5_complex.step")
        chip_gen.visualize_chip_with_points()
    else:
        print("✗ Failed to generate chip")

# Run example
generate_complex_customization()
````

## Summary and Best Practices

### Key Points:

1. **Runtime Configuration**: You can now pass custom electrode parameters directly when creating `IonTrapChipGenerator`:
   ```python
   custom_params = {('region_key', electrode_index): {'parameter': value}}
   chip_gen = IonTrapChipGenerator(custom_electrode_params=custom_params)
   ```

2. **Cumulative Positioning**: Electrode positions are calculated cumulatively, so modifying earlier electrodes affects all subsequent ones.

3. **Parameter Priority**: Individual parameters override defaults. Only specify parameters you want to change.

4. **Spatial Consistency**: The system automatically maintains geometric consistency.

### Recommended Workflow:

1. Start with default parameters and visualize
2. Identify electrodes that need customization
3. Create custom parameter dictionary
4. Generate chip with custom parameters
5. Visualize and verify
6. Export final design

### Parameter Ranges:

- **Electrode width**: 0.040 - 0.120 mm
- **Gap width**: 0.030 - 0.080 mm
- **Electrode depth**: 0.05 - 0.20 mm
- **Electrode height**: 0.5 - 2.0 mm

### Common Use Cases:

- **Loading electrodes**: Wider width, deeper depth
- **RF electrodes**: Standard width, shallow depth
- **Control electrodes**: Narrow width, tight gaps
- **Detection electrodes**: Wider width
- **Gradient designs**: Gradually changing widths
